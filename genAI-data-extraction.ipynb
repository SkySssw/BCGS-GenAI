```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df = df.sort_values(['Year','Company'])
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
int_col = ['Total revenue','Net income','Total assets','Total liabilities', 'Cash Flow from Operating Activities']
df[int_col] = df[int_col].astype('float')
```

```python
# Feature engineering
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net income'].pct_change() * 100
df['Liabilities Change (%)'] = df.groupby(['Company'])['Total liabilities'].pct_change() * 100
df['Total Assets Change (%)'] = df.groupby(['Company'])['Total assets'].pct_change() * 100
df['Profit Margin (%)'] = df['Net income']/df['Total revenue'] * 100
df['Return on Assets(%)'] = df['Net income']/df['Total assets'] * 100
df['Shareholder s equity'] = df['Total assets'] - df['Total liabilities']
df['Return on Equity (%)'] = df['Net income']/df['Shareholder s equity'] * 100
df['Asset turnover ratio'] = df['Total revenue']/df['Total assets']
```

```python
df.sort_values(['Company','Year'])
```

```python
df.to_csv('data_transform.csv')
```
